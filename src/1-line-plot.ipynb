### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 11)  # X values from 0 to 10
y = x ** 2  # Y values as the square of X

# Error handling: Check if x and y are not empty and have the same length
if len(x) > 0 and len(y) > 0 and len(x) == len(y):
    # Create line plot
    plt.plot(x, y, marker='o', linestyle='-', color='b', label="y = x²")

    # Labels and title
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Line Plot of y = x²")
    plt.legend()
    
    # Enable grid
    plt.grid(True)

    # Show plot
    plt.show()
else:
    print("Error: x and y must be non-empty and have the same length.")


```
