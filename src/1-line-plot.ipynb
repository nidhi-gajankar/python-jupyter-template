### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# Error handling: Check if x and y are not empty and have the same length
if x and y and len(x) == len(y):
    plt.plot(x, y, marker='o', linestyle='-', color='b', label="Line Plot")
    
    # Labels and title
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Line Plot Example")
    plt.legend()
    
    # Show plot
    plt.grid(True)
    plt.show()
else:
    print("Error: x and y must be non-empty and of the same length.")

```
