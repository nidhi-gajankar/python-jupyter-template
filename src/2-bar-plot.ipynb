### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'purple']  # Different colors for each bar

# Error handling: Check if categories and values are not empty and have the same length
if len(categories) > 0 and len(values) > 0 and len(categories) == len(values):
    # Create bar plot
    plt.bar(categories, values, color=colors)

    # Labels and title
    plt.xlabel("Categories")
    plt.ylabel("Values")
    plt.title("Bar Plot Example")

    # Show plot
    plt.show()
else:
    print("Error: Categories and values must be non-empty and have the same length.")


```
