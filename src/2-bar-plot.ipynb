### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'purple']  # Different colors for each bar

# Create Bar Plot
plt.bar(categories, values, color=colors)

# Customize Plot
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot Example with Different Colors")

# Show Plot
plt.show()
```
