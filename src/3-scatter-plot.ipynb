### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Error handling: Check if x and y are not empty and have the same length
if len(x) > 0 and len(y) > 0 and len(x) == len(y):
    # Create scatter plot
    plt.scatter(x, y, color='r', marker='o', label="Data Points")

    # Annotate each point with its (x, y) value
    for i in range(len(x)):
        plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='right')

    # Labels and title
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Scatter Plot with Annotations")
    plt.legend()
    
    # Show plot
    plt.show()
else:
    print("Error: x and y must be non-empty and have the same length.")


```
